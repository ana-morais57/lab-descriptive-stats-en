#### Group the employees based on their number of years of employment. What are the average salaries in each group?

#### Find the minimum, mean and the maximum of all numeric columns for each Department.

##### Hint: Use functions from Data Manipulation lesson

```python
# your answer here
num_columns = employee.select_dtypes(include=['number']).columns
department_describe = employee.groupby("Department")[num_columns].agg(["min", "mean", "max"])
department_describe = department_describe.astype(int)
department_describe
```

### Bonus Question

#### For each department, compute the difference between the maximal salary and the minimal salary.

##### * Hint: try using `agg` or `apply` and `lambda`*

```python
# your answer here
department_salary_difference = employee.groupby("Department")["Salary"].agg(lambda x: x.max() - x.min()).reset_index().rename(columns={"Salary": "Salary Difference"})
department_salary_difference.set_index("Department", inplace=True)
department_salary_difference
```
